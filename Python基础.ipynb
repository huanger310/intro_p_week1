# Python基础

## 一、GIT基础
1. git add：把文件添加到暂存区
```
git add file_name 
git add *
```
2. git commit：提交文件,引号中是注释
```
git commit -m 'first commit'
```
3. git push：推送到远程库
```
git push
git push origin master
```
4. git status:查看仓库状态
```
git status
```
5. mkdir:创建目录
```
mkdir huangergit
```
6. pwd：显示当前目录的路径
```
pwd
```
7. git init:把当前的目录变成可以管理的git仓库，生成隐藏.git文件
```
git init
```
8. git diff  XX      查看XX文件修改了那些内容
```
git diff readme.txt
```
9. git log          查看历史记录
```
git log
```
10. git reset  –hard HEAD^回退到上一个版本
```
git reset –hard HEAD~100
```
11. cat XX         查看XX文件内容
```
cat xx
```
12. git reflog       查看历史记录的版本号id
```
git reflog
```
13. git checkout — XX  把XX文件在工作区的修改全部撤销。
```
git checkout — XX
```
14. git rm XX          删除XX文件
```
git rm XX
```
15. git remote add origin xxx 关联一个远程库
```
git remote add origin xxx
```
16. git clone xxx   从远程库中克隆
```
git clone xxx
```
17. git checkout –b dev  创建dev分支 并切换到dev分支上
```
git checkout –b dev
```
18. git branch  查看当前所有的分支
```
git branch
```
19. git checkout master 切换回master分支
```
git checkout master
```
20. git merge dev    在当前的分支上合并dev分支
```
git merge dev
```
21. git branch –d dev 删除dev分支
```
git branch –d dev
```
22. git branch name  创建分支
```
git branch name
```
23. git remote 查看远程库的信息
```
git remote
```
24. git remote –v 查看远程库的详细信息
```
git remote –v
```

## 二、基本代码

1. 创建环境
```
conda create -n intro_p python=3.6
```
2. 进入环境
```
activate intro_p
```
3. 退出环境
```
deactivate intro_p
```
4. 进入jupyter notebook
```
jupyter notebook
```
5. 退出 jupyter notebook
```
两次按Ctrl+C
```
6. 列出全部环境
```
conda env list
```
